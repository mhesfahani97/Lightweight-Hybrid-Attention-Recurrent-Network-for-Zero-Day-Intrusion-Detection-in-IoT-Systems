# <center> libraries </center>

```python
import pandas as pd
import numpy as np
from imblearn.over_sampling import SMOTE
import matplotlib.pyplot as plt
```

# <center> reading ton dataframe </center>

```python
input_path = f'your-path\\dataset\\TON-IoT\\normal-attack\\csv-4\\ton.csv'
ton_df = pd.read_csv(input_path, on_bad_lines="error", low_memory=False)
```

# <center> Extended synthetic sampling </center>

```python
X = ton_df.drop(columns=['type'])
y = ton_df['type']

smote = SMOTE(sampling_strategy={7: 20000}, k_neighbors=5, random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

ton_df = pd.DataFrame(X_resampled, columns=X.columns)
ton_df['type'] = y_resampled
```

```python
X = ton_df.drop(columns=['type'])
y = ton_df['type']

smote = SMOTE(sampling_strategy={3: 20000}, k_neighbors=5, random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

ton_df = pd.DataFrame(X_resampled, columns=X.columns)
ton_df['type'] = y_resampled
```

```python
X = ton_df.drop(columns=['type'])
y = ton_df['type']

smote = SMOTE(sampling_strategy={1: 20000}, k_neighbors=5, random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

ton_df = pd.DataFrame(X_resampled, columns=X.columns)
ton_df['type'] = y_resampled
```

```python
X = ton_df.drop(columns=['type'])
y = ton_df['type']

smote = SMOTE(sampling_strategy={9: 50000}, k_neighbors=5, random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

ton_df = pd.DataFrame(X_resampled, columns=X.columns)
ton_df['type'] = y_resampled
```

# <center> save </center>

```python
ton_path = f'your-path\\dataset\\TON-IoT\\normal-attack\\csv-5\\ton.csv'
ton_df.to_csv(ton_path, index=False)
```
