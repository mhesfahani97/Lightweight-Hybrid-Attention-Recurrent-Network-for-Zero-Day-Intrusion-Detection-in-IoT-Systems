# <center> libraries </center>

```python
import pandas as pd
```

# <center> reading dataframes </center>

```python
ddos_path = f'your-path\\normal-attack\\csv-2\\ddos.csv'
backdoor_path = f'your-path\\normal-attack\\csv-2\\backdoor.csv'
password_path = f'your-path\\normal-attack\\csv-2\\password.csv'
ransomware_path = f'your-path\\normal-attack\\csv-2\\ransomware.csv'
injection_path = f'your-path\\normal-attack\\csv-2\\injection.csv'
scanning_path = f'your-path\\normal-attack\\csv-2\\scanning.csv'
dos_path = f'your-path\\normal-attack\\csv-2\\dos.csv'
mitm_path = f'your-path\\normal-attack\\csv-2\\mitm.csv'
xss_path = f'your-path\\normal-attack\\csv-2\\xss.csv'
normal_path = f'your-path\\normal\\csv-2\\normal.csv'

ddos_df = pd.read_csv(ddos_path, on_bad_lines="error", low_memory=False)
backdoor_df = pd.read_csv(backdoor_path, on_bad_lines="error", low_memory=False)
password_df = pd.read_csv(password_path, on_bad_lines="error", low_memory=False)
ransomware_df = pd.read_csv(ransomware_path, on_bad_lines="error", low_memory=False)
injection_df = pd.read_csv(injection_path, on_bad_lines="error", low_memory=False)
scanning_df = pd.read_csv(scanning_path, on_bad_lines="error", low_memory=False)
dos_df = pd.read_csv(dos_path, on_bad_lines="error", low_memory=False)
mitm_df = pd.read_csv(mitm_path, on_bad_lines="error", low_memory=False)
xss_df = pd.read_csv(xss_path, on_bad_lines="error", low_memory=False)
normal_df = pd.read_csv(normal_path, on_bad_lines="error", low_memory=False)
```

# <center> labeling and selecting type </center>

```python
# ddos

attackers_ip = ["192.168.1.30", "192.168.1.31", "192.168.1.34", "192.168.1.35", "192.168.1.36", "192.168.1.37", "192.168.1.38"]

ddos_df = ddos_df[ddos_df['src_ip'].isin(attackers_ip)].copy()
ddos_df.loc[:, "type"] = 0
ddos_df.loc[:, "label"] = 1
```

```python
# backdoor

attackers_ip = ["192.168.1.33", "192.168.1.37"]

backdoor_df = backdoor_df[backdoor_df['src_ip'].isin(attackers_ip)].copy()
backdoor_df.loc[:, "type"] = 1
backdoor_df.loc[:, "label"] = 1
```

```python
# password

attackers_ip = ["192.168.1.30", "192.168.1.31", "192.168.1.32", "192.168.1.35", "192.168.1.38"]

password_df = password_df[password_df['src_ip'].isin(attackers_ip)].copy()
password_df.loc[:, "type"] = 2
password_df.loc[:, "label"] = 1
```

```python
# ransomware

attackers_ip = ["192.168.1.33", "192.168.1.37"]

ransomware_df = ransomware_df[ransomware_df['src_ip'].isin(attackers_ip)].copy()
ransomware_df.loc[:, "type"] = 3
ransomware_df.loc[:, "label"] = 1
```

```python
# injection

attackers_ip = ["192.168.1.30", "192.168.1.31", "192.168.1.33", "192.168.1.35"]

injection_df = injection_df[injection_df['src_ip'].isin(attackers_ip)].copy()
injection_df.loc[:, "type"] = 4
injection_df.loc[:, "label"] = 1
```

```python
# scanning

attackers_ip = []
for i in range(20, 39):
    attackers_ip.append(f"192.168.1.{i}")

scanning_df = scanning_df[scanning_df['src_ip'].isin(attackers_ip)].copy()
scanning_df.loc[:, "type"] = 5
scanning_df.loc[:, "label"] = 1
```

```python
# dos

attackers_ip = ["192.168.1.30", "192.168.1.31", "192.168.1.39"]

dos_df = dos_df[dos_df['src_ip'].isin(attackers_ip)].copy()
dos_df.loc[:, "type"] = 6
dos_df.loc[:, "label"] = 1
```

```python
# mitm

attackers_ip = ["192.168.1.31", "192.168.1.34"]

mitm_df = mitm_df[mitm_df['src_ip'].isin(attackers_ip)].copy()
mitm_df.loc[:, "type"] = 7
mitm_df.loc[:, "label"] = 1
```

```python
# xss

attackers_ip = ["192.168.1.32", "192.168.1.35", "192.168.1.36", "192.168.1.39"]

xss_df = xss_df[xss_df['src_ip'].isin(attackers_ip)].copy()
xss_df.loc[:, "type"] = 8
xss_df.loc[:, "label"] = 1
```

## <center> normal </center>

```python
normal_df.loc[:, "type"] = 9
normal_df.loc[:, "label"] = 0
```

# <center> drop duplicate </center>

```python
for df in [normal_df, ddos_df, backdoor_df, password_df, ransomware_df, injection_df, scanning_df, dos_df, mitm_df, xss_df]:
    df.drop(columns=["src_ip", "dst_ip", "src_port", "dst_port", "ts"], inplace=True)

for df in [normal_df, ddos_df, backdoor_df, password_df, ransomware_df, injection_df, scanning_df, dos_df, mitm_df, xss_df]:
    print(len(df))

for df in [normal_df, ddos_df, backdoor_df, password_df, ransomware_df, injection_df, scanning_df, dos_df, mitm_df, xss_df]:
    df.drop_duplicates(inplace=True)

print("\n--------------\n")

for df in [normal_df, ddos_df, backdoor_df, password_df, ransomware_df, injection_df, scanning_df, dos_df, mitm_df, xss_df]:
    print(len(df))
```

# <center> randomly choosing samples </center>

```python
ddos_df = (ddos_df.sample(n=20000, random_state=42))
password_df = (password_df.sample(n=20000, random_state=42))
injection_df = (injection_df.sample(n=20000, random_state=42))
scanning_df = (scanning_df.sample(n=20000, random_state=42))
dos_df = (dos_df.sample(n=20000, random_state=42))
xss_df = (xss_df.sample(n=20000, random_state=42))
```

# <center> create ton dataframe </center>

```python
ton_df = pd.concat([ddos_df, backdoor_df, password_df, ransomware_df, injection_df, scanning_df, dos_df, mitm_df, xss_df, normal_df], ignore_index=True)
```

# <center> save </center>

```python
ton_path = f'your-path\\normal-attack\\csv-3\\ton.csv'
ton_df.to_csv(ton_path, index=False)
```
